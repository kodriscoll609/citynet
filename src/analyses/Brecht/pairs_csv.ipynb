```python
# load packages
import re, string
import pandas as pd
import numpy as np
import os
import time
from tqdm.notebook import tqdm
import unidecode

# import from script
from preprocessing_functions import *
```

```python
def process_link(string):
#     pattern_pair = "\((.*?)\)"
    pattern_pair = "\'(.*?)\'"
    pattern_cooc = ": (.*?\.0)"
    
    pair = re.findall(pattern_pair, string)
    cooc = re.findall(pattern_cooc, string)
    
    return pair[0], pair[1], float(cooc[0])
```

```python
string = "('London', 'Copenhagen'): 1530.0"
process_link(string)
```

```python
OUTDIR = "../../output"
FILE = "citypairs_english_20220601.txt"
fp = os.path.join(OUTDIR, FILE)

city_links = []

with open(fp, 'r') as f:
    for line in f.readlines():
        city_links.append(tuple(process_link(line)))
```

```python
df_en = pd.DataFrame(city_links, columns = ['city_A', 'city_B', 'collocation'])
```

```python
df_en.head()
```

```python
file_name = "citypairs_english_20220603.csv"
fp = os.path.join(OUTDIR, file_name)
df_en.to_csv(fp, index = False, sep = ";")
```

```python
df_en.sort_values('collocation', ascending = False)[:20]
```
